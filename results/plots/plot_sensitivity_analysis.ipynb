```python
# Assuming LOCATION_RESULTS is predefined and s_values, w_values, NO_BINS_values are given
LOCATION_RESULTS = ""
s_values = [20, 40, 60]  # Example, replace with actual logic
w_values = ['30S', '1Min', '3Min', '5Min']  # Example, replace with actual logic
NO_BINS_values = [5]  # Example, replace with actual logic

all_rules = []  # To store rules data from all files

# Load and process each file
for s in s_values:
    for w in w_values:
        for NO_BINS in NO_BINS_values:
            filename = f'ALL_{s}_{w}_{NO_BINS}_rules_output.txt'
            filepath = '../resk1000/' + filename
            try:
                rules = pd.read_csv(filepath, sep='#', header=None)
                rules.columns = ['Rule', 'Support', 'Confidence']
                rules.Support = rules.Support.str.replace('SUP: ', "").astype(float)/(2000000/s)
                rules.Confidence = rules.Confidence.str.replace('CONF: ', "").astype(float)
                rules['s'] = s
                rules['w'] = w
                rules['NO_BINS'] = NO_BINS
                all_rules.append(rules)
            except FileNotFoundError:
                print(f"File not found: {filename}")
                continue

# Concatenate all dataframes into one
all_rules_df = pd.concat(all_rules)
```

```python
# Filter rules containing a number between 1 and 31
# pattern = re.compile(r'\b(?:3[01]|[12][0-9]|0?[1-9])\b')
# pattern = re.compile(r'\b(?:3[02-3]|[12][0-9]|0?[1-9])\b')
pattern = re.compile(r'\b(?:0|3[02-3]|[12][0-9]|0?[1-9])\b')
filtered_rules_df = all_rules_df[all_rules_df['Rule'].apply(lambda x: bool(pattern.search(x)))]

# Aggregate the statistics for these filtered rules
aggregated_stats_filtered = filtered_rules_df.groupby('Rule').agg(
    Occurrences=('Rule', 'size'),
    Max_Confidence=('Confidence', 'max'),
    Max_Support=('Support', 'max'),
    Min_Confidence=('Confidence', 'min'),
    Min_Support=('Support', 'min')
).reset_index()

# Find corresponding s, w, NO_BINS for each max and min
for index, row in aggregated_stats_filtered.iterrows():
    rule = row['Rule']
    max_conf_row = filtered_rules_df[(filtered_rules_df['Rule'] == rule) & (filtered_rules_df['Confidence'] == row['Max_Confidence'])].iloc[0]
    max_supp_row = filtered_rules_df[(filtered_rules_df['Rule'] == rule) & (filtered_rules_df['Support'] == row['Max_Support'])].iloc[0]
    min_conf_row = filtered_rules_df[(filtered_rules_df['Rule'] == rule) & (filtered_rules_df['Confidence'] == row['Min_Confidence'])].iloc[0]
    min_supp_row = filtered_rules_df[(filtered_rules_df['Rule'] == rule) & (filtered_rules_df['Support'] == row['Min_Support'])].iloc[0]

    aggregated_stats_filtered.at[index, 'Max_Confidence_w_s_NO_BINS'] = f"{max_conf_row['w']},{max_conf_row['s']},{max_conf_row['NO_BINS']}"
    aggregated_stats_filtered.at[index, 'Max_Support_w_s_NO_BINS'] = f"{max_supp_row['w']},{max_supp_row['s']},{max_supp_row['NO_BINS']}"
    aggregated_stats_filtered.at[index, 'Min_Confidence_w_s_NO_BINS'] = f"{min_conf_row['w']},{min_conf_row['s']},{min_conf_row['NO_BINS']}"
    aggregated_stats_filtered.at[index, 'Min_Support_w_s_NO_BINS'] = f"{min_supp_row['w']},{min_supp_row['s']},{min_supp_row['NO_BINS']}"

# Display the final dataframe
print(aggregated_stats_filtered.sort_values('Occurrences', ascending=False))
```

```python
# Create a directed graph
G = nx.DiGraph()
aggregated_stats_filtered.Max_Confidence = aggregated_stats_filtered.Max_Confidence.astype(float)
aggregated_stats_filtered.Max_Support = aggregated_stats_filtered.Max_Support.astype(float)
# Add edges, confidence, and support as attributes
for index, row in aggregated_stats_filtered.loc[:, ['Rule', 'Max_Confidence', 'Max_Support']][aggregated_stats_filtered.Max_Confidence >= 0.0].iterrows():
    antecedents, consequents = row['Rule'].split(' ==> ')
    antecedents = antecedents.replace(' ', '')
    consequents = consequents.replace(' ', '')
    confidence = row['Max_Confidence']
    support = row['Max_Support']  # Assume we now have support values
    for antecedent in antecedents.split(','):
        for consequent in consequents.split(','):
            G.add_edge(antecedent, consequent, confidence=confidence, support=support)

# Normalize support values for color mapping
supports = [G[u][v]['support'] for u, v in G.edges()]
colors = plt.cm.viridis(np.interp(supports, (min(supports), max(supports)), (0, 1)))

fig, ax = plt.subplots(figsize=(15, 15))
pos = nx.circular_layout(G)
nx.draw_networkx_nodes(G, pos, ax=ax, node_size=2000, node_color='lightblue')
nx.draw_networkx_labels(G, pos, ax=ax)

# Extract confidence values, mapped to a suitable range for edge widths
confidences = [d['confidence'] for _, _, d in G.edges(data=True)]
# Map confidence values to edge widths, e.g., from 0.5 to 5
edge_widths = [0.08 + 4.5 * (conf - min(confidences)) / (max(confidences) - min(confidences)) for conf in confidences]

i=0
for (u, v), color in zip(G.edges(), colors):
    
    start_pos = np.array(pos[u])
    end_pos = np.array(pos[v])
    rad = 0.1  # Adjust curvature radius here if necessary

    arrow = mpatches.FancyArrowPatch(start_pos, end_pos, arrowstyle='-|>, head_width=.2, head_length=1.3', color=color,
                                     connectionstyle=f'arc3,rad={rad}', mutation_scale=20, lw=edge_widths[i])
    ax.add_patch(arrow)
    i += 1

# Setting the aspect of the plot to 'equal' to maintain the scale
ax.set_aspect('equal')
plt.axis('off')

# Add colorbar for support values
sm = plt.cm.ScalarMappable(cmap=plt.cm.viridis, norm=plt.Normalize(vmin=min(supports), vmax=max(supports)))
sm.set_array([])
cbar = plt.colorbar(sm, orientation='vertical', fraction=0.046, pad=0.04)
cbar.set_label('Support')

plt.title("")
plt.show()
```

```python
plt.figure(figsize=(20,20))
G_undirected = G.to_undirected()

# Find all connected components, each component is a set of nodes
connected_components = list(nx.connected_components(G_undirected))

# Filter components to keep only those with 3 or more nodes
large_components = [comp for comp in connected_components if len(comp) >= 20]

# Extract subgraphs from the original graph based on the filtered components
subgraphs = [G.subgraph(comp).copy() for comp in large_components]

# If you need to work with or visualize these subgraphs:
for i, subgraph in enumerate(subgraphs, start=1):
    print(f"Subgraph {i}, Nodes: {subgraph.nodes()}, Edges: {subgraph.edges()}")
    # You can also visualize each subgraph using nx.draw or similar visualization functions
    # For example:
    nx.draw(subgraph, with_labels=True)
```

```python
# Calculating the IQR and determining the range for each signal to exclude outliers
ranges = {}
for signal in signals:
    q25, q75 = np.percentile(dataset_descr[signal], [15, 85])
    iqr = q75 - q25
    lower_bound = q25 - 1.5 * iqr
    upper_bound = q75 + 1.5 * iqr
    ranges[signal] = (lower_bound, upper_bound)

# # Plotting histograms again without outliers
# fig, axs = plt.subplots(len(signals), 1, figsize=(10, 20))

# for i, signal in enumerate(signals):
#     data_to_plot = dataset_descr[signal][(dataset_descr[signal] >= ranges[signal][0]) & (dataset_descr[signal] <= ranges[signal][1])]
#     axs[i].hist(data_to_plot, bins=20, density=True, alpha=0.6, color='b', range=ranges[signal])
#     axs[i].set_title(signal)
#     axs[i].set_ylabel('Probability')
#     axs[i].set_xlabel('Value')

# plt.tight_layout()
# plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add edges with probabilities (you can also add nodes explicitly if needed)
edges_with_probabilities = [('A', 'B', 0.8), ('A', 'C', 0.9), ('B', 'C', 0.7), ('C', 'D', 0.6)]
for u, v, p in edges_with_probabilities:
    G.add_edge(u, v, probability=p)

# Draw the graph
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightblue", font_size=20, font_weight="bold", arrowstyle="->", arrowsize=20)

# Draw edge labels
edge_labels = {(u, v): f'{d["probability"]}' for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.title("Transition Graph with Probabilities")
plt.show()
```
